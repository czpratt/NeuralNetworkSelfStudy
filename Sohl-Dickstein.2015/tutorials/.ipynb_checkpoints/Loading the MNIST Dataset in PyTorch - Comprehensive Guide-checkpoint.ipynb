### How to load MNIST data using Pytorch

even JAX recommends using it since it's really good, and plus this will offload myself from relying on the Tensorflow way of doing things

```python
import torch
from torchvision import datasets, transforms
```

```python
# Define a transform, which to convert the data to tensor and normalize it
# Normalization with these values (mean, std) are specific to MNIST
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Load the training dataset with normalization
train_dataset = datasets.MNIST(root='data', 
                               train=True, 
                               download=True, 
                               transform=transform)

# Load the test dataset with normalization
test_dataset = datasets.MNIST(root='data', 
                              train=False, 
                              download=True, 
                              transform=transform)
```

```python
from torch.utils.data import DataLoader

# Define batch size
batch_size = 64

# Create data loader for the training dataset,
# which allow us to actually work with the data
# making the training data be shuffeled, and it will directly
# from train_dataset
train_loader = DataLoader(dataset=train_dataset, 
                          batch_size=batch_size, 
                          shuffle=True)

# Create data loader for the test dataset
test_loader = DataLoader(dataset=test_dataset, 
                         batch_size=batch_size,
                         shuffle=False)
```

```python
import matplotlib.pyplot as plt

# Function to visualize images
def show_images(images, labels):
    fig, axes = plt.subplots(1, len(images), figsize=(10, 2))
    for img, label, ax in zip(images, labels, axes):
        ax.imshow(img.squeeze(), cmap='gray')
        ax.set_title(f'Label: {label}')
        ax.axis('off')
    plt.show()

# Get a batch of images from the training data loader
# iter() "activates" the training data set so we are ready to recieve data
# next() "supplies" us with a batch of images from the training data set
# you'll only ever do this with one batch at a time
images, labels = next(iter(train_loader))

# Display the first 5 images
show_images(images[:10], labels[:10])
```

Now let's train a simple NN with this data as a way to test out the guide, and for my future understanding

```python
import torch.nn as nn

class SimpleNN(nn.Module):
    ''' 
        Simple Neural Network with two fully connected linear layers, 
        and one hidden layer
        
        Layers:
            fc1: fully connected layer 1: we will be recieving 28x28 pixels in the form
                 of a flattened 784 dimensional vector. then we will be outputting 
                 a 128 dimensional vector, so we are "embedding" 784 pixels into 128
                 dimensions, so we're "squashing" our data into this vector. this is how
                 we create "features" in the network, so the neurons will be learning
                 more properties of the digits during this embedding process so that 
                 they will activate for different reasons.
                 --- analogous to this is Embedding for bigram LLM for example, but 
                     remember that nn.Embedding is a lookup table to find where
                     a character was within the network in order to find out 
                     correlations between the characters/tokens themselves.
                     fc1 is a transformation, which takes a vector of 784
                     and outputs a vector of dim 128, and things are indexed the way 
                     they are in the langauge model examples
                     
            relu: our hidden activation layer, which provides a nonlinearity inbetween
                  the two linear layers
            fc2: fc1 but again, and this time we are going to further squash into 10
            softmax: our final activation function, which normalizes our outputs to be 
                     used to evaluate our loss function
    '''
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128) # (748, 128), 128 is a choice
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 10) # (128, 10), choosing 10 bc there's 10 digits
        self.softmax = nn.LogSoftmax(dim=1) # dim = 1 to normalize across scores of pred nums
        
    def forward(self, x):
        ''' Forward pass, which is going to be handcoded for this simple model '''
        x = x.view(-1, 28 * 28) # flattening for input, will need to see dim size
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return self.softmax(x)

model = SimpleNN()
```

Now let's set up the loss calcutor and the optimizer

```python
criterion = nn.NLLLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
```

```python
epochs = 5
for epoch in range(epochs):
    running_loss = 0.0
    for images, labels in train_loader:
        # Always zero your gradients
        optimizer.zero_grad()
        
        # Forward pass
        outputs = model(images)
        
        # Evaluate the loss based on the forward pass
        loss = criterion(outputs, labels)
        
        # Backward pass
        loss.backward()
        
        # Optimize based on the backward pass
        optimizer.step()
        
        # Accumulate our loss, which is obtaining the loss for each batch
        # that we are iterating over
        running_loss += loss.item()
        
    # when we are done evaluating over the batch, we normalize our running
    # loss by how many batches there are in the training set
    print(f'Epoch {epoch + 1}, Loss: {running_loss / len(train_loader)}')
```

```python
# Evaluate the model
correct = 0
total = 0
# set to no_grad because we aren't training
with torch.no_grad():
    # iterate through all images and labels in our testing data set
    for images, labels in test_loader:
        # get outputs, which we'll use to compare against actual labels
        outputs = model(images)
        
        # Get the probability associated with a number that is largest
        _, predicted = torch.max(outputs, dim=1)
        
        # get the total number of labels there are in the data set as a running total
        # should be 10,000 since we split the data set
        total += labels.size(0)
        
        # for all of the image identifications in a batch,
        # sum up the ones that were correct
        correct += (predicted == labels).sum().item()

# obtain accuracy as a percentage
accuracy = 100 * correct / total
print(f'Accuracy on test set: {accuracy:.2f}%')
```
