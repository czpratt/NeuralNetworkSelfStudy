#### 2) Now that I now how to produce an array of random numbers that is the same size as a 28x28 MNIST digit, how could I iteratively destroy the image using this kind of random sampling? Basically how can I recreate the forward diffusion process from scratch?

```python
# Let's just try something, why not? 
# I'm smart and quick on my feet, there's no need to be scared
import tensorflow as tf  # Used for the data loading pipeline
import tensorflow_datasets as tfds

# I'll simplify this at some point, as it's pretty annoying, but it's
# not super important right now
def get_mnist_dataset(batch_size=64):
    """
    Downloads and prepares the MNIST dataset using tensorflow_datasets,
    and returns an iterator that yields JAX-compatible NumPy arrays.
    
    Returns:
        An iterator that yields batches of (images, labels).
    """
    
    # --- 1. Load the Dataset with TFDS ---
    # as_supervised=True returns (image, label) tuples.
    ds = tfds.load('mnist', split='train', as_supervised=True)
    
    # --- 2. Define the Preprocessing Function ---
    # This function will be mapped over the dataset.
    # TFDS loads images as uint8 [0, 255].
    def normalize_img(image, label):
        # Cast to float
        image = tf.cast(image, tf.float32)
        # Normalize to [-1, 1]
        # (pixel / 127.5) - 1.0
        # (0 / 127.5) - 1.0 = -1.0
        # (255 / 127.5) - 1.0 = 2.0 - 1.0 = 1.0
        image = (image / 127.5) - 1.0
        return image, label

    # --- 3. Build the tf.data Pipeline ---
    ds = ds.map(normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.cache()  # Cache the preprocessed data in memory
    ds = ds.shuffle(1000)  # Shuffle the dataset
    ds = ds.batch(batch_size)  # Batch the data
    ds = ds.prefetch(tf.data.AUTOTUNE)  # Prefetch batches for performance
    
    # --- 4. Convert to NumPy Iterator ---
    # tfds.as_numpy converts the tf.Tensor batches into NumPy arrays,
    # which JAX can consume without any issues.
    # ~ I am guessing where this step differs between JAX and Pytorch
    return tfds.as_numpy(ds)

# --- Example Usage ---
if __name__ == "__main__":
    
    # To run this, you'll need: 
    # pip install jax jaxlib tensorflow tensorflow-datasets
    
    # JAX models are pure functions, so the data pipeline is
    # typically an iterator that yields NumPy arrays.
    
    print("Loading MNIST data with JAX / tensorflow_datasets...")
    train_iter = get_mnist_dataset(batch_size=128)
    
    # Get one batch of data
    # 'images_np' will be a NumPy array of shape [batch_size, 28, 28, 1]
    # 'labels_np' will be a NumPy array of shape [batch_size]
    images_np, labels_np = next(iter(train_iter))
    
    # You can then convert them to JAX arrays for use in a train step
    images_jnp = jnp.array(images_np)
    labels_jnp = jnp.array(labels_np)
    
    print(f"\n--- JAX Data Info ---")
    print(f"Data iterator created successfully.")
    print(f"Batch of images shape: {images_jnp.shape}")
    print(f"Data format: [Batch Size, Height, Width, Channels] (Channel-Last)")
    print(f"Labels shape: {labels_jnp.shape}")
    print(f"Image data type: {images_jnp.dtype}")
    print(f"Image min value: {images_jnp.min():.2f}")
    print(f"Image max value: {images_jnp.max():.2f}")
    print(f"Label data type: {labels_jnp.dtype}")
```

### Next: Let's mimic the exact forward diffusion process, which turns an MNIST digit into white Gaussian noise

```python
fig, axs = plt.subplots(1, 1, figsize=(2,2))
axs.imshow(first_img, cmap='gray')
```

In order to transform this into white Gaussian noise, we can transform it due to the properties of Gaussians:

$\vec{x}(t) = \mu + \mathbb{L}\epsilon = \mu + \mathcal{E}$

where $\mu$ is the mean, $\mathbb{L}$ is the Cholesky factor, which when multiplied by the transpose of itself $\mathbb{L}\mathbb{L}^{\mathsf{T}} = \Sigma$ gives the covariance matrix $\Sigma$. Finally, $\epsilon \sim \mathcal{N}(0, \mathbb{I})$ is a sample from a unity-covariance normal Gaussian that has zero mean, where $\mathbb{I}$ is the identity matrix.

In Sohl-Dickenstein 2015, the mean and covariance matrix are given in the Table App. 1: $\mu = \vec{x}(t-1) \sqrt{1 - \beta_t}$ and $\Sigma = \beta \mathbb{I}$. This means that the Cholesky factor is $\sqrt{\beta}\mathbb{I}$.

Let's create this in our update formula!

```python
# let beta_t be pretty small
# let alpha = sqrt(1 - beta_t)
key = jax.random.key(42)
beta_t, alpha_t = 0.1, jnp.sqrt(1 - beta_t)
mu = first_img * alpha_t
Eps = jnp.sqrt(beta_t) * jnp.ones(first_img.shape) * jax.random.normal(key, shape=(first_img.shape))

x_t1 = mu + Eps
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5,5))
axs[0].imshow(first_img, cmap='gray')
axs[1].imshow(x_t1, cmap='gray')
```

I did it for one image! Now, let's run this through a bunch of steps, and completely obliterate the original image over 10 steps

```python
def forward_process(first_img, beta_t, nsteps, key):
    ''' 
        Transform one image into complete noise 
    
        args: first_img, beta_t, key
        output: list of images that I want to plot, because I want to visualize the 
                the forward diffusion process
    '''
    
    # img_arr will keep track of all of the images because I want to plot them
    # alpha_t contains the diffusion rate step
    # mu is the mean
    # Epsilon contains the random sampling
    img_list = [] # Python lists are mutable, JAX jnp arrays are not
    alpha_t = jnp.sqrt(1 - beta_t)
    for i in range(nsteps):
        '''
            Iterate over nsteps to produce each image,
            Use the previous image to continue the denoising process
        
            We'll make this more efficient later on,
            and make better use of the randomness key
        '''
        # recalculate mean mu and Eps every time with either the first img, 
        # or the next image that's in the array
        if i == 0:
            img_list.append(first_img)
        
        mu = img_list[i] * alpha_t
            
        Eps = jnp.sqrt(beta_t) * jnp.ones(first_img.shape) * jax.random.normal(key, shape=(first_img.shape))
        
        new_img = mu + Eps
                
        img_list.append(new_img)
        
    img_arr = jnp.array(img_list)
    return img_arr
```

```python
first_img = images_jnp[0, :, :, 0]
_beta_t = 0.01
print(int(1/_beta_t))
img_arr = forward_process(first_img, _beta_t, int(1/_beta_t), jax.random.key(42))
```

```python
import matplotlib.pyplot as plt

ncols = 10
fig, axs = plt.subplots(1, ncols, figsize=(20, 5))

for i in range(ncols):
    axs[i].imshow(img_arr[i], cmap='gray')
```

## Let's now repeat the forward diffusion kernel being implemented on 10 images, so this will require changing the function a little bit

```python
# how about we'll give the function images_jnp, and then we can pass
# an argument specifying how many images we want to iterate through
# and apply the kernel to it
def forward_diffusion_many(images, nimages, beta_t, key):
    ''' 
        Applying forward process on many images 
    
        Will be making more efficient later, just wanting to make it work first
    '''
    img_list = [] # this will contain all the images
    alpha_t = jnp.sqrt(1 - beta_t)
    for nimg in range(nimages):
        static_img = images[nimg, :, :, 0]
        simg_list = []
        
        # now apply diffusion kernal on just this image
        for i in range(int(1/beta_t)):
            if i == 0:
                simg_list.append(static_img)
            
            # calculate mu and Eps
            mu = simg_list[i] * alpha_t
            Eps = jnp.sqrt(beta_t) * jnp.ones(static_img.shape) * jax.random.normal(key, shape=(static_img.shape))
            
            new_img = mu + Eps
            simg_list.append(new_img)
    
        # at the end of the static image noising, append to img_list
        simg_arr = jnp.array(simg_list)
        img_list.append(simg_arr)
        
    # at the end of all of the image noising, make it a jnp array
    img_arr = jnp.array(img_list)
    return img_arr
```

```python
nimgs = 5
img_arr = forward_diffusion_many(images_jnp, nimgs, 0.05, jax.random.key(42))
img_arr.shape
```

```python
ncols = 10
fig, axs = plt.subplots(nimgs, ncols, figsize=(20, 10))
for i in range(nimgs):
    for j in range(ncols):
        axs[i][j].imshow(img_arr[i, j, :, :], cmap='gray')
        
        axs[i][j].set_xticklabels([])
        axs[i][j].set_yticklabels([])
```
