```python
import torch
import torch.nn.functional as F
```

```python
words = open('names.txt', 'r').read().split()
```

```python
N = torch.zeros((27, 27), dtype=torch.int32)
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
```

```python
xs, ys = [], []

for w in words[:1]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
        
# then we'll turn these into tensors since we'll use them to build the network with PyTorch
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27, 27), generator=g, requires_grad=True) # remember to set requires_grad
```

```python
# Forward pass
xenc = F.one_hot(xs, num_classes=27).float()
logits = xenc @ W
counts = logits.exp()
probs = counts / counts.sum(1, keepdims=True)
loss = -probs[torch.arange(len(ys)), ys].log().mean()
```

```python
# Backward pass
W.grad = None # remember to always zero your grads
loss.backward()
```

```python
# Update data
W.data += -0.1 * W.grad
```
