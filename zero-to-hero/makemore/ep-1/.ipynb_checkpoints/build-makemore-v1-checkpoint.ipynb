Now we're going to look at bigrams (only two characters) in order to predict what the next character is

We're going to start by just iterating through the first word for two consecutive characters in a row

```python
for w in words[:1]:
    for ch1, ch2 in zip(w, w[1:]):
        print(ch1, ch2)
```

How the above works is the following: w is 'emma' while w[1:] is 'mma'. So what zip does is take two iterators, ch1 and ch2, and match up these two things together until one is shorter than the other, then zip halts.

this is why we see e m , then m m, then m a, then zip() halts!

Now we're going to modify this by including special characters that line up with the start and end of a name because we're already starting to see that certain letters are likely to be at the start or end of a name

We're going to indicate these with start ['<S>'] and end ['<E>'] characters

```python
for w in words[:1]:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        print(ch1, ch2)
```

With these bigram models, we'll start by understanding the frequency of characters by simply counting how frequently the letters appear. We'll save this in a dictionary

We use .get as a dictionary method which will instantiate the value of the bigram to 0 if it hasn't appeared in the dictionar yet, and we'll always increment by 1 otherwise

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
        #print(ch1, ch2)
```

```python
b
```

Just for fun, I'm going to sort this dictionary.

The way this works is that sorted() will take in the items of b in the form (key, tuple), and then we're going to use the key of sorted in order to specify how we're going to order the items. For example we could use key=len in order to specify that we want to sort by the length of the values. Here, we want to sort by the value of item[1], which is the value in the (key, value) pair of b, and then we're going to reverse the order so that the most frequent letters appear first

```python
sorted_dict = dict(sorted(b.items(), key=lambda item: item[1], reverse=True))
sorted_dict
```

There's a couple of things to note here. First is that the bottom row has all zeros, since <E> marks the end of a name and you wouldn't expect anything to follow it afterwards. Also, the second to last column is all 1s, since nothing will begin before the <S> key
    
Let's revise into something more clever in order to account for this happening
 

```python
N = torch.zeros((27, 27), dtype=torch.int32)
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

First what we'll do is make the process of sampling randomly from N, and then converting the row into probabilities, so that this process is way more efficient

Keep in mind that dim=0 refers to the rows of the tensor, where you will be summing up the elements in all of the rows of a tensor in parallel, while dim=1 is ditto but w.r.t. to the column

So in a way it's opposite. Specifying dim=0 means that all the columns in the X row will be summed up, and will become elements in a new row vector. But, if we specify dim=1, then all of the rows in the Y cols will be summed up in a new column vector

```python
a = torch.tensor([[1, 2, 3], [4, 5, 6]])
a
#print(torch.sum(a, dim=0))  # Output: tensor([5, 7, 9])
```

```python
a.sum(dim=0, keepdim=True).shape
```

Now let's do this for our tensor N

We want to have a column vector at the end, which is the sum of all of each element in individual rows

```python
P = N.float()
P.sum(1, keepdim=True).shape # this sums up everything in the rows
```

But remember that we now technically have to divide by P. But we don't necessarily know if we can do the following operation: P / P.sum(1, keepdim=True)

The way to figure out if we can do this kind of operation is called tensor broadcasting

```python
# recall that P.shape is 27, 27
# recall that P.sum .. is 27, 1
# according to Tensor broadcasting, this operation should work
# the P.sum is 'broadcasted' so that we can have element-wise division across all 27 cols in P
P = N.float()
P = P / P.sum(dim=1, keepdim=True)
```

Now let's do the same thing again, but with this new normalized tensor, and we should get the same result

```python
# now let's sample, e.g. 10 names
g = torch.Generator().manual_seed(2147483647)
for _ in range(20):
    ix = 0
    myname = []
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        myname.append(itos[ix])
        if ix == 0: # if we randomly sample a dot, then we need to stop, since nothing will come after it
            break
    print(''.join(myname)) 
```

Let's move onto understanding the probabilities of bigrams occuring in our model

```python
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        print(f'{ch1}{ch2}: {prob:.4f}')
```

We now want a way to be able to assess how good the model is at actually producing names that look like they'd work. Perhaps we could assign a single number to assess how good the model is at coming up with names?

This usually requires looking at the likelihood, which is the product of the probabilities of the outcomes in your model.

What we want is that the product of probabilities to be high, because that's how we are concatenating our names together from strings. But multiplying numbers between 0 and 1 isn't that useful, so we'll look at the log likelihood in order to assess how good the model is at putting together names

Note that in print(f'{var=}') that you'll also be print out whatever the var is that you passed in

```python
log_likelihood = 0.0
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        print(f'{ch1}{ch2}: {prob:.4f}, {logprob:.4f}')
        
print(f'{log_likelihood=}')
```

Remember though that we want to define a loss function; semantically, when the loss function is minimized, this is a good thing. So what that means is that we really want to look at the negative log likelihood!

```python
nll = 0.0
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        nll -= logprob
        print(f'{ch1}{ch2}: {prob:.4f}, {logprob:.4f}')
        
print(f'{nll=}')
```

Now we have the intuition for this dataset and how we can assess how good of a model it is: When we are trying to minimize the loss function, we are aiming to send nll to zero

What's conventional to use is a type of normalized loss function. The intuition still holds, though; the lower that this averaged loss function is, the better

```python
nll = 0.0
n = 0
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        nll -= logprob 
        n += 1
        print(f'{ch1}{ch2}: {prob:.4f}, {logprob:.4f}')
        

norm_nll = nll/n        
print(f'{nll=}')
print(f'{norm_nll=}')
```

Now the real question is: What parameters do we need to adjust in order to minimize norm_nll? This will be answered once we start working with neural networks to more accurately predict things

Notice that when we add a unnatural letter at the end of my name, like x, we see that it is basically impossible for x to follow s, so this makes our loss function go to infinity, which is obviously not desired

```python
nll = 0.0
n = 0
#for w in words:
for w in ['chrisx']:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        nll -= logprob 
        n += 1
        print(f'{ch1}{ch2}: {prob:.4f}, {logprob:.4f}')
        

norm_nll = nll/n        
print(f'{nll=}')
print(f'{norm_nll=}')
```

We can address this by smoothing out our model of these potential infinities---one way to do this is to just add a 1 time occurrence in our probability matrix in order to not necessarily make the entire distribution more uniform, but to keep the peaks of the distribution without having our loss go to infinity

```python
P = (N+1).float()
P /= P.sum(dim=1, keepdims=True)
```

```python
nll = 0.0
n = 0
#for w in words:
for w in ['chrisx']:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        nll -= logprob 
        n += 1
        print(f'{ch1}{ch2}: {prob:.4f}, {logprob:.4f}')
        

norm_nll = nll/n        
print(f'{nll=}')
print(f'{norm_nll=}')
```

Now it's just incredibly unlikely that a x will follow an s, and this makes our loss function large but not infinity, which is good
