```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

### Building the dataset

```python
words = open('../names.txt', 'r').read().splitlines()
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
```

```python
block_size = 3 # context length: adjust how many characters will be used to predict the next one
X, Y = [], []
for w in words[:5]:
    #print(f'word: {w}')
    context = [0] * block_size
    
    for ch in w + '.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        print(''.join(itos[i] for i in context), '--->', itos[ix])
        context = context[1:] + [ix]
        
X = torch.tensor(X)
Y = torch.tensor(Y)
```

Initialization

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn(27, 2)
W1 = torch.randn((6, 100))
b1 = torch.randn(100)
W2 = torch.randn((100, 27))
b2 = torch.randn(27)
parameters = [C, W1, b1, W2, b2]
for p in parameters:
    p.requires_grad = True
    p.grad = None
```

1) Re-embed on the forward pass since we'll adjust C since it is our feature vector

```python
for _ in range(100):
    # Forward pass
    emb = C[X]
    h = torch.tanh(emb.view(-1, emb.shape[1]*emb.shape[2]) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y)
    
    # Backward pass
    loss.backward()
    
    # Update by training step and calculated grad
    for p in parameters:
        p.data += -0.1 * p.grad

print(f'loss: {loss.item():.4f}')
```

### You must zero your gradients before you do your backward passes!

```python
block_size = 3 # context length: adjust how many characters will be used to predict the next one
X, Y = [], []
for w in words:
    context = [0] * block_size    
    for ch in w + '.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        context = context[1:] + [ix]
        
X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn(27, 2, generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
for p in parameters:
    p.requires_grad = True
```

```python
for i in range(100):
    # Forward pass
    ix = torch.randint(0, X.shape[0], (32,))
    emb = C[X[ix]]
    h = torch.tanh(emb.view(-1, emb.shape[1]*emb.shape[2]) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y[ix])
    
    # Backward pass
    for p in parameters:
        p.grad = None
    loss.backward()
    
    # Update by training step and calculated grad
    for p in parameters:
        p.data += -0.1 * p.grad

print(f'loss at step {i+1}: {loss.item():.4f}')
```

Punchline: It is much better to get an approximate gradient and iterate it over many steps (what we're doing when we're training on batches) rather than trying to always calculate an exact gradient but not be able to iterate it over many steps (what we were doing when we were not using batches and it took take a long time)

## Now we're going to move on a bit and think about what's a good way that we can determine the learning rate--we had just chosen it kind of arbitrarily, but what's a good way of rigorously being able to find it? or maybe not rigorously, but how to find a ballpark estimate of what's good without needing to try each possibility

It doesn't make a whole lot of sense to step through the learning rates in a linear way, since we're not working with linear counts or probabilities, so what we can do is be able to iterate through different possible learning rates through their exponentials!

```python
lre = torch.linspace(-3, 0, 1000)
#lrs_ = torch.exp(lre)
lrs = 10**lre
#plt.plot(lrs_) # we're literally looking at the exponential space of learning rates
#plt.plot(lre) # as opposed to just the linear learning rates
#plt.plot(lrs2) # i wonder if there's a difference betwen using these two? 
```

```python
g = torch.Generator().manual_seed(2147483647)
lrs = 10**(torch.linspace(-3, 0, 1000))
C = torch.randn(27, 2, generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
for p in parameters:
    p.requires_grad = True
    p.grad = None
```

In order to find what the optimal value for the learning rate is, we need to step through the training and save the loss and learning rate at a particular training step. We can plot all of this data to visualize it. Then we can find what learning rate and loss was minimal, which helps for this optimization scheme!

```python
# keep track of sats
lri, lossi = [], []

for i in range(1000):
    # Forward pass
    ix = torch.randint(0, X.shape[0], (32,))
    emb = C[X[ix]]
    h = torch.tanh(emb.view(-1, emb.shape[1]*emb.shape[2]) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y[ix])
    
    # Backward pass
    for p in parameters:
        p.grad = None
    loss.backward()
    
    # Update by training step and calculated grad
    lr = lrs[i]
    for p in parameters:
        p.data += -lr * p.grad
    lri.append(lr)
    lossi.append(loss.item())
    
print(f'loss at step {i+1}: {loss.item():.4f}')
```

```python
plt.plot(lri, lossi)
plt.xlabel(f'learning rate')
plt.ylabel(f'loss')
plt.show()
```

Now we're going to talk about the ins and outs of potentially overfitting your data. As your model grows in number of parameters (we're currently at 3k params, but remember that LLaMa is at what 400B?) you risk overfitting your training set. so what's common in the field to do is to split up your dataset into three different batches: train, dev and test. train is for training your hyperparameters, dev is for (?) and test is where you're actually testing your model out to see how good it can do.

But what can happen is that if you have a ton of hyperparameters, and then you train it excessively on your training set, the loss can basically go to zero because it is essentially memorizing your training data and even though it looks really good, it's just memorization, and it won't be able to predict that good.

so we split them into the splits--dev is the validation set to see how good it is and validate whether things need to be changed or not, and the test split is used at the very very end to really test things out to see if things were good or not. you use the test split very very sparingly because you don't want your model to memorize this part of the split, as it's supposed to be the 'final exam' of the model

### Building the different datasets

```python
def build_dataset(words):
    block_size = 3
    X, Y = [], []
    
    for w in words:
        context = [0] * block_size
        for ch in w + '.':
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X, Y

# now we'll split the full dataset?
import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
```

### Now we're going to reset the network and we'll be training specifically on this network

```python
g = torch.Generator().manual_seed(2147483647)
lrs = 10**(torch.linspace(-3, 0, 1000))
C = torch.randn(27, 2, generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
for p in parameters:
    p.requires_grad = True
    p.grad = None
```

#### Now we'll be training specifically on Xtr and Ytr, and we'll be evaluating how well our network was trained on Xdev and Ydev

```python
lossi = []
tsteps = 10_000
for i in range(tsteps):
    # Forward pass
    ix = torch.randint(0, Xtr.shape[0], (32,), generator=g)
    emb = C[Xtr[ix]]
    h = torch.tanh(emb.view(-1, emb.shape[1]*emb.shape[2]) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix])
    
    # Backward pass
    for p in parameters:
        p.grad = None
    loss.backward()
    
    # Update by training step and calculated grad
    lr = 0.01 # adjust to 0.01 for the learning rate decay at the end of training
    for p in parameters:
        p.data += -lr * p.grad
    if i % 1000 == 0:
        print(f'step {i} out of {tsteps}: {loss}')
    lossi.append(loss.item())
    
print(f'loss at step {i+1}: {loss.item():.4f}')
```

## Now let's really formalize the hell out of this training step

```python
def build_dataset(words):
    block_size = 3
    X, Y = [], []
    
    for w in words:
        context = [0] * block_size
        for ch in w + '.':
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X, Y

# now we'll split the full dataset?
import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn(27, 10, generator=g)
nneurons = 200
W1 = torch.randn((30, nneurons), generator=g)
b1 = torch.randn(nneurons, generator=g)
W2 = torch.randn((nneurons, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
batch_size = 32
print(f'total parameters: {sum(p.nelement() for p in parameters)}')
for p in parameters:
    p.requires_grad = True
    p.grad = None
```

```python
lossi = []
stepi = []
```

```python
for i in range(200_000):
    # mini-batch construction
    ix = torch.randint(0, Xtr.shape[0], (batch_size,))
    
    # forward pass
    emb = C[Xtr[ix]]
    h = torch.tanh(emb.view(-1, emb.shape[1]*emb.shape[2]) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix])
    
    # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()
    
    # update
    lr = 0.1 if i < 100_000 else 0.01
    for p in parameters:
        p.data += -lr * p.grad
    if i % 10_000 == 0:
        print(f'loss at step {i}: {loss:.4f}')
    lossi.append(loss.log10().item())
    stepi.append(i)
    
print(f'-----\nfinal loss: {10**(lossi[-1])}')
```

### Now we are going to sample from the model to see how good we are doing

```python
g = torch.Generator().manual_seed(2147483647+10)
block_size = 3 # this should be initialized earlier

for _ in range(20):
    out = []
    context = [0] * block_size
    
    while True:
        emb = C[torch.tensor([context])] # (1, contextsize)
        h = torch.tanh(emb.view(1, -1) @ W1 + b1)
        logits = h @ W2 + b2
        probs = F.softmax(logits, dim=1)
        ix = torch.multinomial(probs, num_samples=1, generator=g).item()
        context = context[1:] + [ix]
        out.append(ix)
        if ix == 0:
            break
            
    print(''.join(itos[i] for i in out))
```
